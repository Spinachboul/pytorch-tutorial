```python
import torch
torch.__version__
```

```python
# scalars
scalar = torch.tensor(7)
scalar
```

```python
# checking the dimensions of scalar
scalar.ndim
```

```python
# retrieve the number from the scalar
scalar.item()
```

```python
# vector: a single dimension tensor but can contain many numbers
vector = torch.tensor([7, 7]) # a tensor but with multiple values
vector
```

```python
vector.ndim
```

```python
# check shape of vector
vector.shape
```

```python
# matrix
MATRIX = torch.tensor([
    [7,8],
    [9, 10]
])
MATRIX
```

```python
# checking dimensions
MATRIX.ndim
```

```python
# shape of matrix
MATRIX.shape
```
